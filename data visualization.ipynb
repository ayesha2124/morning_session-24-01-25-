```python
#bivariate

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, label="Line", color="blue",linestyle="--",linewidth=2,marker="o")  
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Line Plot")
plt.show()
```

```python
#bivariate

import matplotlib.pyplot as plt
x = ['jan', 'feb', 'mar', 'april', 'may','june']
y = [11000, 4000, 6000, 8000, 10000,110000]
plt.plot(x, y,color="green",linestyle="dashed",marker="o")  
plt.xlabel("months(1-6)")
plt.ylabel("Savings amount")
plt.title("Growth rate")
plt.show()
```

```python
import matplotlib.pyplot as plt
#bivariate
x = ['a', 'b', 'c', 'd', 'e','f']
y = [110000, 40000, 100000, 80000, 10000,110000]
plt.bar(x, y,color="maroon",edgecolor="lightgray")  
plt.xlabel("categories")
plt.ylabel("Savings amount")
plt.title("Growth rate")
plt.show()
```

```python
import matplotlib.pyplot as plt
#bivariate
x = ['a', 'b', 'c', 'd', 'e','f']
y = [110000, 40000, 100000, 80000, 10000,110000]
plt.bar(x, y,color=["coral", "turquoise", "peachpuff"],edgecolor="lightgray")  
plt.xlabel("categories")
plt.ylabel("Savings amount")
plt.title("Growth rate")
plt.show()
```

```python
#histogram
```

```python
import matplotlib.pyplot as plt
#bivariate
ages=[11,15,17,20,25,29,30,40,45,48,50,58,60,68,70,80,90]#bins=def the intervals for grouping the data
y = [10,15,20,25,30]
plt.hist(ages,bins=y,color="skyblue",edgecolor="black",width=4)  
plt.xlabel("age distribution of customers")
plt.ylabel("age ranges")
plt.title("frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
#bivariate
ages=[11,15,17,20,25,29,30,40,45,48,50,58,60,68,70,80,90]#bins=def the intervals for grouping the data
y = [10,15,20,25,30]
plt.hist(ages,bins=y,color="maroon",edgecolor="black")  
plt.xlabel("age distribution of customers")
plt.ylabel("age ranges")
plt.title("frequency")
plt.grid(True) #by default it is false 
plt.show()
```

```python
#piechart
```

```python
import matplotlib.pyplot as plt

languages = ['C', 'Python', 'Java', 'HTML', 'JS']
percentages = [10, 15, 20, 25, 30]

plt.pie(percentages, labels=languages, colors=["yellow", 'green', 'brown', 'skyblue', 'lavender'], autopct='%1.1f%%',startangle=90)
plt.title("Programming Languages Distribution")
plt.legend()
plt.show()
```

```python
#bivariate

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.scatter(x, y,color="blue",s=100,edgecolor="black")  
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("scatter Plot")
plt.show()
```
