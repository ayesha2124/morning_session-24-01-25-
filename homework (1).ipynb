```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'Sales': [1500, 1800, 2200, 1900, 2500, 2800],
        'OrderQuantity': [10, 15, 20, 12, 25, 30]}
df = pd.DataFrame(data)
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'Sales': [1500, 1800, 2200, 1900, 2500, 2800],
        'OrderQuantity': [10, 15, 20, 12, 25, 30]}
df = pd.DataFrame(data)
plt.hist(df['OrderQuantity'], bins=5, edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
        'Math': [85, 92, 78, 95, 88, 70, 65, 98, 80, 82],
        'Science': [90, 88, 85, 92, 75, 82, 78, 95, 88, 91],
        'English': [78, 85, 92, 80, 88, 95, 70, 82, 90, 75]}
df = pd.DataFrame(data)
subjects = ['Math', 'Science', 'English']
avg_marks = df[subjects].mean()
plt.bar(subjects, avg_marks, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Average Student Marks Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.show()
sns.boxplot(data=df[subjects])
plt.title('Distribution of Student Marks Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```
